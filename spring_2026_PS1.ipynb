# Analytical Derivation of the Log-Likelihood Function

The Poisson probability mass function (PMF) for a single observation $y_i$ with parameter $\lambda$ is given by:

$P(Y_i = y_i | \lambda) = \frac{\lambda^{y_i} e^{-\lambda}}{y_i!}$

For a sample of $n$ independent and identically distributed (i.i.d.) observations $y_1, y_2, \ldots, y_n$, the likelihood function $L(\lambda; y_1, \ldots, y_n)$ is the product of the individual PMFs:

$L(\lambda; y_1, \ldots, y_n) = \prod_{i=1}^{n} \frac{\lambda^{y_i} e^{-\lambda}}{y_i!}$

To simplify calculations, we typically work with the log-likelihood function, denoted as $\ell(\lambda; y_1, \ldots, y_n)$. Taking the natural logarithm of the likelihood function:

$\ell(\lambda; y_1, \ldots, y_n) = \ln \left( \prod_{i=1}^{n} \frac{\lambda^{y_i} e^{-\lambda}}{y_i!} \right)$

Using the properties of logarithms ($\ln(AB) = \ln A + \ln B$ and $\ln(A/B) = \ln A - \ln B$, and $\ln(A^B) = B \ln A$):

$\ell(\lambda; y_1, \ldots, y_n) = \sum_{i=1}^{n} \ln \left( \frac{\lambda^{y_i} e^{-\lambda}}{y_i!} \right)$

$\ell(\lambda; y_1, \ldots, y_n) = \sum_{i=1}^{n} (\ln(\lambda^{y_i}) + \ln(e^{-\lambda}) - \ln(y_i!))$

$\ell(\lambda; y_1, \ldots, y_n) = \sum_{i=1}^{n} (y_i \ln(\lambda) - \lambda - \ln(y_i!))$

Finally, we can separate the sum:

$\ell(\lambda; y_1, \ldots, y_n) = \left( \sum_{i=1}^{n} y_i \right) \ln(\lambda) - n\lambda - \sum_{i=1}^{n} \ln(y_i!)$

This is the analytical form of the log-likelihood function for the Poisson distribution.

### Analytical Derivation of the Score Contribution

The Score function for a single observation $y_i$, denoted as $S(y_i; \lambda)$, is the first derivative of the log-likelihood function for a single observation with respect to $\lambda$.

The log-likelihood for a single observation $y_i$ is:

$\ell(\lambda; y_i) = y_i \ln(\lambda) - \lambda - \ln(y_i!)$

Now, we take the derivative with respect to $\lambda$:

$S(y_i; \lambda) = \frac{\partial \ell(\lambda; y_i)}{\partial \lambda}$

$S(y_i; \lambda) = \frac{\partial}{\partial \lambda} (y_i \ln(\lambda) - \lambda - \ln(y_i!))$

$S(y_i; \lambda) = y_i \frac{\partial}{\partial \lambda} (\ln(\lambda)) - \frac{\partial}{\partial \lambda} (\lambda) - \frac{\partial}{\partial \lambda} (\ln(y_i!))$

Since $\frac{\partial}{\partial \lambda} (\ln(\lambda)) = \frac{1}{\lambda}$, $\frac{\partial}{\partial \lambda} (\lambda) = 1$, and $\ln(y_i!)$ is a constant with respect to $\lambda$ (so its derivative is 0):

$S(y_i; \lambda) = y_i \left(\frac{1}{\lambda}\right) - 1 - 0$

$S(y_i; \lambda) = \frac{y_i}{\lambda} - 1$

This is the analytical form of the Score contribution for a single observation from the Poisson distribution.

### Analytical Derivation of the Hessian Matrix

The Hessian matrix (or simply the Hessian in this univariate case) for a single observation $y_i$, denoted as $H(y_i; \lambda)$, is the second derivative of the log-likelihood function with respect to $\lambda$. Alternatively, it is the first derivative of the Score function with respect to $\lambda$.

From the previous step, the Score contribution for a single observation $y_i$ is:

$S(y_i; \lambda) = \frac{y_i}{\lambda} - 1$

Now, we take the derivative of the Score function with respect to $\lambda$:

$H(y_i; \lambda) = \frac{\partial S(y_i; \lambda)}{\partial \lambda}$

$H(y_i; \lambda) = \frac{\partial}{\partial \lambda} \left( \frac{y_i}{\lambda} - 1 \right)$

$H(y_i; \lambda) = \frac{\partial}{\partial \lambda} (y_i \lambda^{-1}) - \frac{\partial}{\partial \lambda} (1)$

Using the power rule for differentiation $(\frac{\partial}{\partial x} (x^n) = nx^{n-1})$:

$H(y_i; \lambda) = y_i (-1 \cdot \lambda^{-2}) - 0$

$H(y_i; \lambda) = -\frac{y_i}{\lambda^2}$

This is the analytical form of the Hessian contribution for a single observation from the Poisson distribution.

For a sample of $n$ independent and identically distributed (i.i.d.) observations, the total Hessian is the sum of individual Hessians:

$H(\lambda; y_1, \ldots, y_n) = \sum_{i=1}^{n} H(y_i; \lambda) = \sum_{i=1}^{n} \left( -\frac{y_i}{\lambda^2} \right) = -\frac{1}{\lambda^2} \sum_{i=1}^{n} y_i$

```python
def pois_log_lik(lam, y):
    """
    Poisson log-likelihood.
    Parameters
    ----------
    lam : array-like or scalar Poisson mean(s)
    y : array-like Observed counts
    Returns
    -------
    float
    Log-likelihood value
    """
    # Ensure y is a numpy array for element-wise operations
    y = np.asarray(y)

    # Calculate the log-likelihood based on the derived formula:
    # ell(lambda; y_1, ..., y_n) = (sum(y_i)) * ln(lambda) - n*lambda - sum(ln(y_i!))
    # Or equivalently, sum(y_i * ln(lambda) - lambda - ln(y_i!))

    # Using gammaln(y+1) for ln(y!) which is numerically stable
    log_lik = np.sum(y * np.log(lam) - lam - sp.gammaln(y + 1))

    return log_lik
```

```python
from scipy.optimize import minimize

# --- 1. Simulate n=100 observations from a Poisson distribution with lambda=15 ---
n = 100
true_lambda = 15
simulated_data = np.random.poisson(true_lambda, n)

print(f"Simulated {n} observations from Poisson(lambda={true_lambda}).")
print(f"Sample mean of simulated data: {np.mean(simulated_data):.2f}")

# --- 2. Define the negative log-likelihood function for optimization ---
def neg_log_lik(lam, y):
    # Our pois_log_lik function returns the log-likelihood, so we negate it
    return -pois_log_lik(lam, y)

# --- 3. Find the MLE using scipy.optimize.minimize ---

# Initial guess for lambda (e.g., sample mean of the data)
initial_guess = np.mean(simulated_data)

# Minimize the negative log-likelihood
# We need to ensure lambda (lam) is positive, so we set bounds.
# The 'L-BFGS-B' method is generally good for bounded optimization.
result = minimize(neg_log_lik, initial_guess, args=(simulated_data,),
                    method='L-BFGS-B', bounds=[(0.001, None)]) # lambda must be > 0

mle_lambda = result.x[0] if isinstance(result.x, np.ndarray) else result.x

print(f"\nMLE of lambda: {mle_lambda:.4f}")
print(f"Optimization successful: {result.success}")
print(f"Message from optimizer: {result.message}")
```

```python
def pois_se(lambda_hat, y):
    """
    Calculates the standard error of the Poisson MLE.

    Parameters
    ----------
    lambda_hat : float
        The Maximum Likelihood Estimator (MLE) of lambda.
    y : array-like
        Observed counts from the Poisson distribution.

    Returns
    -------
    float
        The standard error of the Poisson MLE.
    """
    n = len(y) # Number of observations
    se = np.sqrt(lambda_hat / n)
    return se

# Example usage with the previously simulated data and MLE
standard_error_mle = pois_se(mle_lambda, simulated_data)
print(f"Standard error of MLE(lambda): {standard_error_mle:.4f}")
```
